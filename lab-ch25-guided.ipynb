## Part 6: Reflection (10 min)

Answer each question in **3-5 sentences**. Connect your answers to concepts from Chapter 25.

### Q1: What worked well with AI-assisted development?

*Think about: speed of scaffolding, code quality, design suggestions, handling of repetitive tasks.*

**Your answer:**

The initial setup was fast. Claude Code scaffolded the full Next.js project with a nav, hero, projects section, skills section, gallery, and contact page in one shot, which would have taken me a long time to do from scratch. It also handled the repetitive parts well, like generating project cards with consistent formatting and wiring up the dark/light mode toggle. Most of my time went toward decisions about content and layout rather than writing code.

### Q2: What did you have to fix manually?

*Think about: layout bugs, incorrect assumptions, styling mismatches, missing features, hallucinated APIs.*

**Your answer:**

There were a few things that needed fixing. The initial bio and reflection answers came out sounding like AI wrote them, with flowery language and em-dashes everywhere, so I had to go back and rewrite those to sound more normal. The grid layout left blank black areas when filtering projects by a technology that only appeared in one lab, which needed a CSS fix. Claude also included my personal website link in the contact section without me asking, and I had to have that removed. Images were also placed in the wrong folder initially and needed to be moved to public/ to load correctly.

### Q3: How does AI-assisted development compare to writing code entirely yourself?

*Think about: the HADI loop from Chapter 25 (Hypothesis, Action, Data, Insight). Did you follow this pattern? Where did you need domain knowledge that the AI lacked?*

**Your answer:**

Writing it myself would have taken much longer, especially getting the initial Next.js structure set up. The HADI loop moved faster because I could see results right away and iterate from there rather than starting from nothing. The AI did not understand the economics context though, like why the Double ML 401k result is still worth showing even though it is a null finding. I had to explain what projects to include and why, which is the part that actually required my own knowledge.

### Q4: Prompt engineering lessons

*What makes a good prompt for Claude Code? What makes a bad one? Give one specific example of a prompt you improved.*

**Your answer:**

Specific prompts work much better than vague ones. Early on I gave broad instructions and got output that technically worked but did not match what I wanted, like a bio that sounded nothing like how I would describe myself. When I provided my actual resume and gave concrete style direction, the output was much closer on the first try. The main lesson is that the more context you give upfront, the less you have to correct afterward.

### Q5: If you were hiring a junior data scientist, would their portfolio site influence your evaluation? Why or why not?

*Connect to the "signaling" concept from labor economics — what does a portfolio signal that a resume alone does not?*

**Your answer:**

Yes, it would matter. A resume lists what someone has done but a portfolio shows how they think about it and whether they can communicate findings to someone outside their field. In signaling terms a portfolio takes real effort to put together, so it filters out people who just list skills they do not actually have. For a data science role, seeing how someone frames a null result like the 401k finding tells you more about their judgment than a GPA does. Still, I think ability matters most. With generative AI almost anyone can have an impressive portoflio website, but not everyone has developed the correct intuition to figure out the best applications of these tools.
